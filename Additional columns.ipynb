# Additional columns were created to refine clustering output

# Columns such as watch_frequency, watch_pattern, Monday-Sunday, 0-23, etc

```python
import pandas as pd
from scipy.stats import variation
```

```python
df=pd.read_csv("final_for_cluster.csv",index_col=0)
```

```python
df
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'dow'].split(","))
    if y==7:
        df.loc[j,'watch_frequency']='Daily watcher'
    elif y>=4 :
        df.loc[j,'watch_frequency']='Frequent watcher'
    else:
        df.loc[j,'watch_frequency']='Medium watcher'
```

```python
for j in range(0,len(df)):
        df.loc[j,'no_genre_watched']=len(df.loc[j,'genres'].split(","))
df 
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'genres'].split(","))
    z=[]
    for i in range(0, y):
        x=df.loc[j,'genres'].split(",")[i].split(":")
        z.append(int(x[1]))
    
    if df.loc[j,'no_genre_watched']>1:
        df.loc[j,'variance_in_genre']=variation(z)
    else:
        df.loc[j,'variance_in_genre']=0
df  
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'tod'].split(","))
    morning=0
    evening=0
    night=0
    for i in range(0, y):
        x=df.loc[j,'tod'].split(",")[i].split(":")
        if (15>=int(x[0])>=8):
            morning=morning+(int(x[1]))
        elif (21>=int(x[0])>=16):
            evening=evening+(int(x[1]))
        else:
            night=night+(int(x[1]))
    a=max(morning,evening,night)
    if (a==morning):
        
    #if ((morning > evening>night) or (morning>night>evening)):
        df.loc[j,'watch_pattern']='morning'
    elif (a==evening):
    #elif ((evening>morning> night)or(evening>night>morning)):
        df.loc[j,'watch_pattern']='evening'
    else:
        df.loc[j,'watch_pattern']='night'
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'tod'].split(","))
    morning=0
    evening=0
    night=0
    total=0
    for i in range(0, y):
        x=df.loc[j,'tod'].split(",")[i].split(":")
        if (15>=int(x[0])>=8):
            morning=morning+(int(x[1]))
        elif (21>=int(x[0])>=16):
            evening=evening+(int(x[1]))
        else:
            night=night+(int(x[1]))
    total=morning+evening+night
    df.loc[j,'morning_watch']=(morning/total)*100
    df.loc[j,'evening_watch']=(evening/total)*100
    df.loc[j,'night_watch']=(night/total)*100
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'dow'].split(","))
    z=[]
    k=[]
    a=0
    for i in range(0, y):
        x=df.loc[j,'dow'].split(",")[i].split(":")
        z.append(int(x[1]))
        k.append((x[0]))
        a=a+int(x[1])

    for i in range(0,len(k)):
        df.loc[j,k[i]]=(z[i]/a)*100  
df  
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'tod'].split(","))
    z=[]
    k=[]
    a=0
    for i in range(0, y):
        x=df.loc[j,'tod'].split(",")[i].split(":")
        z.append(int(x[1]))
        k.append((x[0]))
        a=a+int(x[1])

    for i in range(0,len(k)):
        df.loc[j,k[i]]=(z[i]/a)*100  
df 
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'genres'].split(","))
    k=[]
    a=0
    for i in range(0, y):
        x=df.loc[j,'genres'].split(",")[i].split(":")
        k.append((x[0]))
        a=a+int(x[1])
    df.loc[j,'mean_genre_time']=(a/len(k))
df 
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'titles'].split(","))
    k=[]
    a=0
    for i in range(0, y):
        x=df.loc[j,'titles'].split(",")[i].rsplit(":",1)
        k.append((x[0]))
        a=a+int(x[1])
    df.loc[j,'mean_title_time']=(a/len(k))
df 
```

```python
for j in range(0,len(df)):
    y= len(df.loc[j,'cities'].split(","))
    k=[]
    a=0
    for i in range(0, y):
        x=df.loc[j,'cities'].split(",")[i].split(":")
        k.append((x[0]))
        a=a+int(x[1])
    df.loc[j,'mean_city_time']=(a/len(k))
df 
```
